```python
#14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature',
#'Humidity' and Perform the following tasks using Matplotlib:
#1. Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis  for 'Temperature' and right y-axis for 'Humidity')
#2. Label the x-axis as 'Date'.
#3. Set the title of the plot as 'Temperature and Humidity Over Time'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Create a time-series dataset
dates = pd.date_range('2023-01-01', periods=100)
df = pd.DataFrame({
 'Date': dates,
 'Temperature': np.random.randint(10, 30, size=100),
 'Humidity': np.random.randint(30, 70, size=100)
})
# Plot the data
fig, ax1 = plt.subplots()
# Plot temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], color='red', 
label='Temperature')
ax1.set_ylabel('Temperature (°C)', color='red')
# Create a twin y-axis for humidity
ax2 = ax1.twinx()
# Plot humidity on the right y-axis
ax2.plot(df['Date'], df['Humidity'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
# Set the x-axis label and title
plt.xlabel('Date')
plt.title('Temperature and Humidity Over Time')
# Show the plot
plt.show()
```

```python
#15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
#1. Plot a histogram of the data with 30 bins.
#2. Overlay a line plot representing the normal distribution's probability density function (PDF).
#3. Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
#4. Set the title of the plot as 'Histogram with PDF Overlay'
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.5, color='skyblue', edgecolor='black', label='Histogram')

# Overlay a line plot representing the normal distribution's probability density function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 1000)
p = np.exp(-0.5 * ((x - np.mean(data)) / np.std(data))**2) / (np.std(data) * np.sqrt(2 * np.pi))
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title of the plot as 'Histogram with PDF Overlay'
plt.title('Histogram with PDF Overlay')

# Display legend
plt.legend()

# Show plot
plt.show()
```

```python
#17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), 
#add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Create two random arrays
x = np.random.rand(100)
y = np.random.rand(100)
# Define a function to determine the quadrant of a point
def quadrant(x, y):
 if x > 0 and y > 0:
  return 1
 elif x < 0 and y > 0:
  return 2
 elif x < 0 and y < 0:
  return 3
 else:
  return 4
# Create a new column in the DataFrame indicating the quadrant of each point
df = pd.DataFrame({'x': x, 'y': y, 'quadrant': np.vectorize(quadrant)
(x, y)})
# Create the scatter plot
sns.scatterplot(data=df, x='x', y='y', hue='quadrant', legend='brief')
# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Quadrant-wise Scatter Plot')
plt.show()
```

```python
#18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and 
#set the title as 'Sine Wave Function'
```

```python
import numpy as np
from bokeh.plotting import figure, show
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)
p = figure(title="Sine Wave Function", y_range=(-1.1, 1.1))
p.grid.grid_line_color = None
p.line(x, y, line_width=2)
p.xaxis.axis_label = 'Angle (radians)'
p.yaxis.axis_label = 'Sine'
show(p)
```

```python
#19 Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on 
#their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
categories = list('ABCDEFGHIJ')
data = {'category': np.random.choice(categories, size=20),
 'value': np.random.randint(0, 100, size=20)}
source = ColumnDataSource(data=data)
p = figure(title="Random Categorical Bar Chart", y_range=(0, 100))
p.vbar(x='category', top='value', width=0.5, color='blue', 
source=source)
hover = p.select(type=HoverTool)
hover.tooltips = [
 ("Category", "@category"),
 ("Value", "@value"),
]
p.xaxis.axis_label = 'Category'
p.yaxis.axis_label = 'Value'
show(p)
```
